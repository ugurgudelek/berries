In this notebook, i try to show some statistic about finance data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("input/raw_data/spy.csv")
```

```python
# insert percent change column into main dataframe
percentage = df.adjusted_close.pct_change(periods=1)
percentage.iloc[0] = 0.0
df['percent_change'] = percentage
```

```python
shifted = df.adjusted_close.shift(periods=1)
shifted[0] = 0.0
diff = df.adjusted_close - shifted
diff[0] = 0.0
df['diff'] = diff
```

```python
change_thr = 0.0035
distribution = (df.percent_change.loc[df.percent_change < -change_thr].count(),
                df.percent_change.loc[np.logical_and(df.percent_change <= change_thr,df.percent_change >= -change_thr)].count(),
                df.percent_change.loc[df.percent_change > change_thr].count())
```

```python
distribution
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python

df['percent_change_tanh'] = df.percent_change.apply(np.tanh) 
df['percent_change_sigmoid'] = df.percent_change.apply(sigmoid) 
```

```python
df['diff_sigmoid'] = df['diff'].apply(sigmoid)
df['diff_tanh'] = df['diff'].apply(np.tanh) 
```
