```python
from nptdms import TdmsFile
from scipy.signal import savgol_filter
```

```python
data = TdmsFile('../../spectrogram-analysis/kanallar/accx_kanal1.tdms').as_dataframe()
data.rename(columns={"/'Accy'/'Untitled'": "acc"}, inplace=True)

# apply smoothing filter
data['filtered'] = savgol_filter(data['acc'], window_length=51, polyorder=2)
```

```python
data
```

```python
import plotly.graph_objects as go

import numpy as np



# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scattergl(
        x = data.index[51200*2+0:51200*2+439*2],
        y = data['acc'][51200*2+0:51200*2+439*2],
        mode = 'markers+lines',
        marker = dict(
            line = dict(
                width = 1,
                color = 'DarkSlateGrey')
        )
    )
)
fig.add_trace(
    go.Scattergl(
        x = data.index[51200*2+0:51200*2+439*2],
        y = data['filtered'][51200*2+0:51200*2+439*2],
        mode = 'lines',
        marker = dict(
            line = dict(
                width = 1)
        )
    )
)

fig.show()
```

```python
def rpm2numpoints(rpm=7000, sampling_freq=51200):
    """
    Return the number of data points for 1 revolution.
    Number of teeth = 3
    """
    return sampling_freq/(rpm/60)
```
