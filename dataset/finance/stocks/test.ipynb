```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('stocks_sample.csv', index_col=0)
df
```

```python
def is_max(x):
    return True if x.max() == x[-1] else False
def is_min(x):
    return True if x.min() == x[-1] else False



df['max_28'] = df['adjusted_close'].rolling(28).apply(is_max)
df['min_28'] = df['adjusted_close'].rolling(28).apply(is_min)

df['label_down'] = df['min_28']
df['label_up'] = df['max_28']
df['label_hold'] = 0.0
df.loc[(df['label_down'] != 1.0) & (df['label_up'] != 1.0), 'label_hold'] = 1.0
```
