```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('load_training_loss.csv')
df['Load Training Loss'] = df['Value'].copy()
df['Load Validation Loss'] = pd.read_csv('load_validation_loss.csv')['Value']
df['PV Training Loss'] = pd.read_csv('pv_training_loss.csv')['Value']
df['PV Validation Loss'] = pd.read_csv('pv_validation_loss.csv')['Value']

df['Epoch'] = df['Step'].copy()
df.drop(['Step','Wall time', 'Value'],inplace=True, axis=1)
df
```

```python
df['Load Training Loss'].iloc[-10:].apply(np.sqrt)
```

```python
ax = df.iloc[:50].plot(x='Epoch', figsize=(10,7))
ax.set_ylabel('Mean Squared Error (MSE)')
```

```python
def plot_graph(path):
    df = pd.read_csv(path)
    df['mse'] = df['Value'].copy()
    df['mae'] = df['Value'].apply(np.sqrt)
    df.plot('Step',['mse','mae'])
```

```python
plot_graph('load_training_loss.csv')
```

```python
plot_graph('load_validation_loss.csv')
```

```python
plot_graph('pv_training_loss.csv')
```

```python
plot_graph('pv_validation_loss.csv')
```
