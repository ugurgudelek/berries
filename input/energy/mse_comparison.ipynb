```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('load.csv')
```

```python
actual,forecast = df.actual, df.forecast
```

```python
actual = norm(actual)
forecast = norm(forecast)
```

```python
def norm(x):
    return (x-x.min())/(x.max()-x.min())
```

```python
def mse(y,yhat):
    return sum((y-yhat)**2)/len(y)
```

```python
mse(actual, forecast)
```

```python
res_df = pd.read_csv('load_result.csv', index_col=0)
res_df = res_df[['actual', 'forecast', 'y0', 'yhat0']]
res_df['yhat0_shifted'] = res_df['yhat0'].shift(periods=96)
res_df = res_df.dropna(axis=0)
```

```python
res_df['y0_norm'] = norm(res_df['y0'])
res_df['yhat0_norm'] = norm(res_df['yhat0'])
res_df['yhat0_shifted_norm'] = norm(res_df['yhat0_shifted'])
```

```python
mse(res_df['y0_norm'],res_df['yhat0_norm'])
```

```python
mse(res_df['y0_norm'], res_df['yhat0_shifted_norm'])
```

```python
mse(res_df['y0'],res_df['yhat0'])
```

```python
mse(res_df['y0'],res_df['yhat0_shifted'])
```
