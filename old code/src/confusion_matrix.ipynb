```python
unique_dates = predictions.Date.unique()
```

```python
max_df = pd.DataFrame(columns=predictions.columns)
# append only positive max samples for any given dates
for date in unique_dates:
    try:
        max_pos_idx = predictions.loc[(predictions.Date == date) & (predictions.Prediction > 0)].Prediction.idxmax()
        max_df = max_df.append(predictions.iloc[max_pos_idx])
    except:
        continue
# append all negative samples
max_df = max_df.append(predictions.loc[predictions.Prediction < 0])
```

```python
y_true = max_df.Actual > 0
y_pred = max_df.Prediction > 0
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
(tn, fp, fn, tp)
```
