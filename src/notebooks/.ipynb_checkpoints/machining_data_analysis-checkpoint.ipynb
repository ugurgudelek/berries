```python
from nptdms import TdmsFile
from scipy.signal import savgol_filter
import plotly.graph_objects as go

import numpy as np
```

```python
def rpm2numpoints(rpm=7000, sampling_freq=51200):
    """
    Return the number of data points for 1 revolution.
    Number of teeth = 3
    """
    return int(sampling_freq/(rpm/60))
```

```python
RPM = 7000
SAMPLING_FREQ = 51200
data = TdmsFile('D:/machining/spectrogram-analysis/kanallar/accx_kanal2.tdms').as_dataframe()
data.rename(columns={"/'Accy'/'Untitled'": "acc"}, inplace=True)
```

```python
RPM = 2653.93
SAMPLING_FREQ = 12800
data = TdmsFile('D:/machining/wear_data/raw/2/acc15120.tdms').as_dataframe()
data.rename(columns={"/'Acc'/'Untitled'": "acc"}, inplace=True)
```

```python
NUM_POINTS_1_CYCLE = rpm2numpoints(rpm=RPM, sampling_freq=SAMPLING_FREQ)
print(f"NUM_POINTS_1_CYCLE:{NUM_POINTS_1_CYCLE}")

# take absolute
data['abs'] = np.abs(data['acc'])
# apply smoothing filter
data['filtered'] = savgol_filter(data['acc'], 
                                 window_length=51,
                                 polyorder=2)
```

```python
START_IDX = 51200*3
NUM_CYCLE = 2


# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scattergl(
        x = data.index[START_IDX:START_IDX+NUM_POINTS_1_CYCLE*NUM_CYCLE],
        y = data['acc'][START_IDX:START_IDX+NUM_POINTS_1_CYCLE*NUM_CYCLE],
        mode = 'markers+lines',
        marker = dict(
            line = dict(
                width = 1,
                color = 'DarkSlateGrey')
        )
    )
)
fig.add_trace(
    go.Scattergl(
        x = data.index[START_IDX:START_IDX+NUM_POINTS_1_CYCLE*NUM_CYCLE],
        y = data['filtered'][START_IDX:START_IDX+NUM_POINTS_1_CYCLE*NUM_CYCLE],
        mode = 'lines',
        marker = dict(
            line = dict(
                width = 1)
        )
    )
)

fig.show()
```
