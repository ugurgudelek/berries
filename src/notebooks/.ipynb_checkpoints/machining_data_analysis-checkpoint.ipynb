```python
from nptdms import TdmsFile
import numpy as np
import pandas as pd

from scipy import signal
from scipy.signal import butter, lfilter, freqz, savgol_filter

import plotly.offline as py
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
py.init_notebook_mode(connected=True)

import matplotlib.pyplot as plt
%matplotlib inline

import scaleogram as scg

PLOT_TEMPLATE = 'simple_white'
```

```python
class VibrationData():
    def __init__(self, path, rpm, sampling_freq):
        self.path = path
        self.rpm = rpm
        self.sampling_freq = sampling_freq
        # read raw tdms data
        self.reading:pd.DataFrame = TdmsFile(path).as_dataframe()
        self.reading = self.reading.rename(columns={self.reading.columns[0]: "data"})
        # add time column
        self.reading['time'] = np.arange(self.reading.shape[0])*(1/sampling_freq)
        
        # Lowpass filter requirements.
        LOWPASS_ORDER = 2     
        CUTOFF = 2500  # desired cutoff frequency of the filter, Hz
        SAVGOL_WINDOW_LEN = int(self.num_points_in_one_period/6)
        SAVGOL_WINDOW_LEN = SAVGOL_WINDOW_LEN if SAVGOL_WINDOW_LEN %2 == 1 else SAVGOL_WINDOW_LEN+1

        # apply savgol filter
        self.reading['savgol'] = savgol_filter(self.reading['data'], 
                                        window_length=SAVGOL_WINDOW_LEN,
                                        polyorder=1)
        # apply lowpass filter
        self.reading['lowpass'] = self.butter_lowpass_filter(self.reading['data'], CUTOFF, self.sampling_freq, LOWPASS_ORDER)

        self.reading['lowpass+savgol'] = savgol_filter(self.reading['lowpass'], 
                                        window_length=SAVGOL_WINDOW_LEN,
                                        polyorder=1)

    @staticmethod
    def _make_trace(data, colname, mode='lines'):
        return go.Scattergl(name=colname,
                            x=data['time'],
                            y=data[colname],
                            mode=mode,
                            marker=dict(line=dict(width=1)))

    def make_subset(self, start_idx):
        return self.reading.iloc[start_idx:start_idx+self.num_points_in_one_period*NUM_CYCLE, :]

    def plot(self, subset):
        fig = go.Figure(data=(self._make_trace(data=subset, colname='data',            mode='markers+lines'),
                              self._make_trace(data=subset, colname='lowpass',         mode='lines'),
                              self._make_trace(data=subset, colname='savgol',          mode='lines'),
                              self._make_trace(data=subset, colname='lowpass+savgol',  mode='lines')),
                       layout = go.Layout(xaxis={'title': 'Time(sec)'},
                                              yaxis={'title': 'Amplitude'},
                                         template=PLOT_TEMPLATE))
        
        # add vertical line per period
        for i in range(NUM_CYCLE):
            fig.add_shape(
               # Line Vertical
               dict(
                   type="line",
                   x0=subset['time'].iloc[self.num_points_in_one_period*i],
                   y0=subset['data'].max(),
                   x1=subset['time'].iloc[self.num_points_in_one_period*i],
                   y1=subset['data'].min(),
                   line=dict(
                       color="RoyalBlue",
                       width=3,
                       dash="dashdot"
                   )
            ))
        fig.show()

    @staticmethod
    def fft(data, sampling_freq, bound=(0, 2500), plot = False):
        Fs=sampling_freq  # sampling freq
        Ts=1/Fs  # sampling interval
        t=data['time']  # time vector

        n=data.shape[0]  # data lenght
        k=np.arange(n)
        T=n/Fs
        freq=k/T

        S=data['data'].values
        S_fft=np.fft.fft(S)


        freq=freq[:n//2]  # one side frequency range
        freq=freq[freq < bound[1]]  # cut unnecessary frequencies

        S_fft=S_fft.real[:freq.shape[0]] / n  # fft computing and normalization
        S_fft=np.abs(S_fft)  # y-axis symetric correction

        fft_data=S_fft
        fft_freqs=freq

        if plot:
            fig=go.Figure(data = [go.Scattergl(x=fft_freqs, y=fft_data, name='fft', mode='markers+lines')],
                           layout = go.Layout(xaxis={'title': 'Freq(Hz)'},
                                              yaxis={'title': 'Amplitude'},
                                             template=PLOT_TEMPLATE))
            fig.show()
        # return fft_data, fft_freqs,fig

    @staticmethod
    def spectrogram(data,sampling_freq, bound = (0, 2500), plot = False):


        spect_freqs, spect_times, spectrogram_data=signal.spectrogram(data['data'], sampling_freq)
        spect_freqs=spect_freqs[spect_freqs < bound[1]]  # cut unnecessary frequencies
        spect_times=spect_times
        spectrogram_data=spectrogram_data[:spect_freqs.shape[0], :]
        
        if plot:
            fig=go.Figure(data = [go.Heatmap(z=10*np.log10(spectrogram_data),
                                             x=spect_times,
                                             y=spect_freqs,
                                             name='spectrogram',
                                            hoverongaps=False,
                                            hovertemplate='X: %{x:.4f}h <br>Y: %{y} <br>Z: %{z}',)],
                           layout = go.Layout(xaxis={'title': 'Time(sec)'},
                                              yaxis={'title': 'Freq(Hz)'},
                                              template=PLOT_TEMPLATE),
                         )
            py.iplot(fig)

        # return spect_freqs, spect_times, spectrogram_data
    
    @property
    def num_points_in_one_period(self):
        """
        Return the number of data points for 1 revolution.
        Number of teeth = 3
        """
        return int(self.sampling_freq/(self.rpm/60))
        
    @staticmethod
    def butter_lowpass_filter(data, cutoff, fs, order=5):
        # butter_lowpass
        nyq = 0.5 * fs
        normal_cutoff = cutoff / nyq
        b, a = butter(order, normal_cutoff, btype='low', analog=False)
        y = lfilter(b, a, data)
        return y
    
    @staticmethod
    def wavelet(data):
        wavelet = 'cmor0.3-1.5'

        # scales = np.logspace(1.0, 1.5, num=200, dtype=np.int32)
        scales = np.arange(1, 200, 1)
        ax = scg.cws(time=data['time'].values, signal=data['data'].values, scales=scales,
                     spectrum='power',
                     wavelet=wavelet,
                     figsize=(12,6), ylabel="Period [seconds]", xlabel='Time [seconds]')
        return ax
```

```python
NUM_CYCLE = 3
# read data
# vib = VibrationData(path='D:/machining/spectrogram-analysis/kanallar/accy_kanal3.tdms', rpm=9000, sampling_freq=51200)
vib = VibrationData(path='D:/machining/wear_data/raw/1/acc7000.tdms', rpm=7000, sampling_freq=12800)

subset = vib.make_subset(start_idx=vib.sampling_freq*2)

# plot data
vib.plot(subset)

# plot fft
vib.fft(data=subset,
        sampling_freq=vib.sampling_freq, 
        plot=True)

# plot spectrogram
vib.spectrogram(data=subset,
                sampling_freq=vib.sampling_freq,
                plot=True)

# plot wavelet
ax = vib.wavelet(data=subset)
for i in range(NUM_CYCLE):
    ax.axvline(x=subset['time'].values[i*vib.num_points_in_one_period], ymin=0.1, ymax=0.9, c='w')
```
