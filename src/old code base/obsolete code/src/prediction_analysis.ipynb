```python
import pandas as pd
from preprocessing import *
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
data = get_last_saved_data()
history = pd.read_pickle("../result/history_dont_delete.pickle")
```

```python
history.keys()
```

```python
predictions = np.array(history['prediction'])
predictions = predictions.reshape(predictions.shape[0])
predictions = pd.DataFrame(predictions)

test_labels = data["test_labels"].ix[:,0]
test_names  = data['test_names'].ix[:,0]
test_dates = data['test_dates'].ix[:,0]
```

```python
predictions.hist(bins=100)
test_labels.hist(bins=100)
```

```python
result = pd.DataFrame()
result['name'] = test_names
result['date'] = test_dates
result['label'] = test_labels
result['prediction'] = predictions
```

```python
result.head()
```

```python
result['disc_label'] = result.label.apply(lambda x: int((x+1)*(1.5)))
result['disc_pred'] = result.prediction.apply(lambda x: int((x+1)*(1.5)))
```

```python
result.head(n=50)
```

```python
real_buy = result[result.disc_label == 2]
real_sell = result[result.disc_label == 0]

total_buy_sgn_count = real_buy.count()
true_buy_sgn_count = np.array([real_buy.disc_pred >= 2]).sum() #1307
wrong_buy_sgn_count = np.array([real_buy.disc_pred == 0]).sum()  #37 / 2564

true_sell_sgn_count = np.array([real_sell.disc_pred == 0]).sum() #1241
wrong_sell_sgn_count = np.array([real_sell.disc_pred >= 2]).sum() #55 / 2389
```

```python
(real_buy.name.count(), true_buy_sgn_count,wrong_buy_sgn_count),(real_sell.name.count(),true_sell_sgn_count,wrong_sell_sgn_count)
```

```python
(np.array([result.disc_pred == result.disc_label]).sum(),result.shape[0])
```

```python
ratio = 0.38
ll = result.label[result.label < -ratio]
ml = result.label[np.logical_and(result.label >= -ratio,result.label <= ratio)]
rl = result.label[result.label > ratio]

lp = result.prediction[result.prediction < -ratio]
mp = result.prediction[np.logical_and(result.prediction >= -ratio,result.prediction <= ratio)]
rp = result.prediction[result.prediction > ratio]
```

```python
ll.hist()
ml.hist()
rl.hist()
lp.hist()
mp.hist()
rp.hist()
```

```python
result['index']=np.array(result.index)
```

```python
result['p_experimental'] = result.prediction.apply(lambda x:x**5)
```

```python
result.describe()
```

```python
result_wo_hold = result[~np.logical_and(result.label >= -ratio,result.label <= ratio)]
```
